```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('/content/fake_superstore_data.csv')
```

```python
df
```

1. Find total sales and average quantity for each region

```python
region_stats = df.groupby('Region')[['Sales', 'Quantity']].agg({'Sales': 'sum', 'Quantity': 'mean'})
region_stats
```

2. For each category, compute standard deviation and CV of sales

```python

category_stats = df.groupby('Category')['Sales'].agg(['mean', 'std'])
category_stats['CV'] = category_stats['std'] / category_stats['mean']
category_stats
```

3. Create column high_discount = True if discount > average of that segment

```python
segment_avg_discount = df.groupby('Segment')['Discount'].transform('mean')
df['high_discount'] = df['Discount'] > segment_avg_discount
segment_avg_discount.head(3)
```

4.Filter sub-categories where total sales > 10,000

```python
subcat_sales = df.groupby('Sub-Category')['Sales'].sum()
high_sales_subcats = subcat_sales[subcat_sales > 10000]
high_sales_subcats
```
