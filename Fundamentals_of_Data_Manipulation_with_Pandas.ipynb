```python
import pandas as pd

Population_series = pd.Series(
    [150, 200, 20, 70, 15, 34],
    index=["Turkey", "Indonesia", "Pakistan", "Nigeria", "Egypt", "Bangladesh"],
    name="Population (Millions)"
)
Population_series



```

```python
df = pd.DataFrame({
    "Name": ["Ali", "Sara", "Hamza", "Ayesha", "Zara"],
    "Department": ["Sales", "HR", "IT", "Finance", "Marketing"],
    "Salary": [70000, 85000, 60000, 90000, 75000],
    "Experience": [2, 5, 3, 7, 4]
})
df
```

```python
df["Bonus"] = df["Salary"] * 0.15
df
```

```python
experienced = df[df["Experience"] > 3]
experienced

```

```python
sorted_df = df.sort_values(by="Salary", ascending=False)
sorted_df
```
